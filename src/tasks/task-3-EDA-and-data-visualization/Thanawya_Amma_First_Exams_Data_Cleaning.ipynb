# **Let's Split the Data Depending on Student Specialization**

- **We have two Main Specialization : Scientific and literary**

- **And Scientific Major splits to two Sub-Specialization : Math and Science**

###**So let's Split Columns depending on the Specializations : Scientific and literary**

```python
df.columns
```

**The diffrence between Scientific and literary is in the Subjects (Columns)**



**We have main subjects (columns) that whole students study :**

 `'arabic'`, `'first_foreign_lang'`,`'religion'`, `'altarbiat_alwatania'`,
`'economics_statistics'`,`'second_foreign_lang'`

> Note `'second_foreign_lang'` vary from student to another, one can studey french and other can study german for example

```python
Main_Subjects = ['arabic', 'first_foreign_lang','religion', 'altarbiat_alwatania', 'economics_statistics','second_foreign_lang']
```

**Now Let's define Each Specializations Subjects :**

**For the literary "ادبي" Specializations we have :**

`'history'`,`'geography'`, `'philosophy'` , `'psychology'`

```python
Literary_Subjects = ['history','geography', 'philosophy' , 'psychology']
```

```python
Main_Scinctific_Subjects = ['chemistry' , 'physics']
```

**For the Science Sub-Specialization we have:**
